```
### sentences

sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```
#Initialize vocabulary size

vocabulary_size = 10000
```

```
#one hot representation

from tensorflow.keras.preprocessing.text import one_hot
onehot = [one_hot(words,vocabulary_size)for words in sent ]
print(onehot[0])

print(onehot)
```

```
#Word Embedding

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```
import numpy as np
```

```
sentence_length = 8
embedding_docs = pad_sequences(onehot,padding='pre',maxlen=sentence_length) #in the 'post' it will add zeros after the docs
#print(embedding_docs[0])
print(embedding_docs)
```

```
dim = 10
```

```
model = Sequential()
model.add(Embedding(vocabulary_size,10,input_length=sentence_length))
model.compile('adam','mse')
```

```
model.summary()
```

```
print(model.predict(embedding_docs))
```
